# data cleaning

# dataset 1

##### Remove column 

```python
df1 = df1.drop('FIRST PAYMENT DATE')
df1 = df1.drop('MATURITY DATE')
df1 = df1.drop('CHANNEL')
df1 = df1.drop('PROPERTY STATE')
df1 = df1.drop('SELLER NAME')
df1 = df1.drop('POSTAL CODE')
df1 = df1.drop('SERVICER NAME')
df1 = df1.drop('Pre-HARP LOAN SEQUENCE NUMBER')
```

```python
from pyspark.sql import functions as F
```

##### 1 CREDIT SCORE

```python
df1 = df1.filter(df1['CREDIT SCORE'] != 9999)
```

##### 3 FIRST TIME HOMEBUYER FLAG

```python
df1 = df1.filter(df1['FIRST TIME HOMEBUYER FLAG'] != '9')
update_func3 = (F.when(F.col('FIRST TIME HOMEBUYER FLAG') == "Y", 1)
                .otherwise(0))
df1 = df1.withColumn('FIRST TIME HOMEBUYER FLAG', update_func3)
```

##### 5 METROPOLITAN STATISTICAL AREA (MSA) OR METROPOLITAN DIVISION

```python
update_func4 = (F.when(F.col('METROPOLITAN STATISTICAL AREA (MSA) OR METROPOLITAN DIVISION')>0,1)
               .otherwise(0))
df1 = df1.withColumn('METROPOLITAN STATISTICAL AREA (MSA) OR METROPOLITAN DIVISION', update_func4)
```

##### 6 MORTGAGE INSURANCE PERCENTAGE (MI %)

```python
df1 = df1.filter(df1['MORTGAGE INSURANCE PERCENTAGE (MI %)'] != '999')
update_func5 = (F.when(F.col('MORTGAGE INSURANCE PERCENTAGE (MI %)') < 1 , 0)
                .otherwise(F.col('MORTGAGE INSURANCE PERCENTAGE (MI %)')))
df1 = df1.withColumn('MORTGAGE INSURANCE PERCENTAGE (MI %)', update_func5)
```

##### 7 NUMBER OF UNITS

```python
df1 = df1.filter(df1['NUMBER OF UNITS'] != 99)
```

##### 8 OCCUPANCY STATUS

```python
import pyspark.sql.functions as F
df1 = df1.filter(df1['OCCUPANCY STATUS'] != '9')

#OCCUPANCY STATUS: change to primary and investment column;
status_dct = {'P':'PRIMARY', 'I':'INVESTMENT'}
for key, item in status_dct.items():
    def status(code):
        if  code == key: 
            return 1
        else:
            return 0
    udf_status_func = F.udf(status, IntegerType())
    df1 = df1.withColumn(item, udf_status_func("OCCUPANCY STATUS"))
```

```python
df1 = df1.drop('OCCUPANCY STATUS')
```

##### 9 ORIGINAL COMBINED LOAN-TO-VALUE (CLTV)

```python
df1 = df1.filter(df1['ORIGINAL COMBINED LOAN-TO-VALUE (CLTV)'] != '999')
```

##### 10 ORIGINAL DEBT-TO-INCOME (DTI) RATIO

```python
df1 = df1.filter(df1['ORIGINAL DEBT-TO-INCOME (DTI) RATIO'] != 999)
```

##### 12 ORIGINAL LOAN-TO-VALUE (LTV): remove 999

```python
df1 = df1.filter(df1['ORIGINAL LOAN-TO-VALUE (LTV)'] != 999)
```

##### 15 PREPAYMENT PENALTY MORTGAGE (PPM) FLAG 


```python
update_func26 = (F.when(F.col('PREPAYMENT PENALTY MORTGAGE (PPM) FLAG') == "Y", 1)
                .otherwise(0))
df1= df1.withColumn('PREPAYMENT PENALTY MORTGAGE (PPM) FLAG', update_func26)
```

##### 16 PRODUCT TYPE

```python
update_func27 = (F.when(F.col('PRODUCT TYPE') == "FRM", 1)
                .otherwise(0))
df1= df1.withColumn('PRODUCT TYPE', update_func27)
```

##### 18 PROPERTY TYPE 

```python
df1 = df1.filter(df1["PROPERTY TYPE"] != '99')
property_type = {'CO':'CONDO', 'PU':'PUD','MH':'MANUFACTURED', 'SF':'SINGLE-FAMILY'}
for key, item in property_type.items():
    def property(code):
        if code == key:
            return 1
        else:
            return 0
    udf_pt_func = F.udf(property, IntegerType())
    df1 = df1.withColumn(item, udf_pt_func("PROPERTY TYPE"))
```

```python
df1 = df1.drop('PROPERTY TYPE')
```

##### 21 LOAN PURPOSE

```python
df1 = df1.filter(df1["LOAN PURPOSE"] != "9")
loan_purpose = {'P':'PURCHASE', 'C':'REFINANCE_CASH_OUT','N':'REFINANCE_NO_CASH_OUT'}
for key, item in loan_purpose.items():
    def purpose(code):
        if code == key:
            return 1
        else:
            return 0
    udf_lp_func = F.udf(purpose, IntegerType())
    df1 = df1.withColumn(item, udf_lp_func("LOAN PURPOSE"))
```

```python
df1 = df1.drop('LOAN PURPOSE')
```

##### 23 NUMBER OF BORROWERS

```python
df1 = df1.filter(df1["NUMBER OF BORROWERS"] != 99)
#column value = 1 when borrower = 1, otherwise = 0 when borrower > 1
update_func23 = (F.when(F.col('NUMBER OF BORROWERS') == "01", 1)
                .otherwise(0))
df1= df1.withColumn('NUMBER OF BORROWERS', update_func23)
```

##### 26 SUPER CONFORMING FLAG

```python
update_func24 = (F.when(F.col('SUPER CONFORMING FLAG') == "Y", 1)
                .otherwise(0))
df1= df1.withColumn('SUPER CONFORMING FLAG', update_func24)
```

##### 28 PROGRAM INDICATOR

```python
update_func25 = (F.when(F.col('PROGRAM INDICATOR') == "H", 1)
                .otherwise(0))
df1= df1.withColumn('PROGRAM INDICATOR', update_func25)
```
