```python
import pyspark.sql.functions as F
df1 = df1.filter(df1['OCCUPANCY STATUS'] != '9')

#OCCUPANCY STATUS: change to primary and investment column;
status_dct = {'P':'PRIMARY', 'I':'INVESTMENT'}
for key, item in status_dct.items():
    def status(code):
        if  code == key: 
            return 1
        else:
            return 0
    udf_status_func = F.udf(status, IntegerType())
    df1 = df1.withColumn(item, udf_status_func("OCCUPANCY STATUS"))
```

```python
def pretty_print_linear(coefs, names = None, sort = False):
    if names == None:
        names = ["X%s" % x for x in range(len(coefs))]
    lst = zip(coefs, names)
    if sort:
        lst = sorted(lst,  key = lambda x:-np.abs(x[0]))
    return " + ".join("%s * %s" % (round(coef, 3), name)
                                   for coef, name in lst)
```

```python
def test(models, data, iterations = 10):
    results = {}
    for i in models:
        r2_train = []
        r2_test = []
        for j in range(iterations):
            X_train, X_test, y_train, y_test = train_test_split(X, 
                                                                Y, 
                                                                test_size= 0.2)
            r2_test.append(metrics.r2_score(y_test,
                                            models[i].fit(X_train, 
                                                         y_train).predict(X_test)))
            r2_train.append(metrics.r2_score(y_train, 
                                             models[i].fit(X_train, 
                                                          y_train).predict(X_train)))
        results[i] = [np.mean(r2_train), np.mean(r2_test)]
    return pd.DataFrame(results)
```

```python
import matplotlib.pyplot as plts
```

```python
from pyspark.sql.functions import isnan, when, count, col, struct
```

```python
from pyspark.sql.functions import max as sparkMax
```

```python
from pyspark.sql.functions import min as sparkMin
```

```python
df1.filter(df1['CREDIT SCORE'] != 9999).count()
```

```python
df1.filter(df1['CREDIT SCORE'] == 9999).count()
```

```python
df1.filter(df1['Credit Score'] != 9999).groupby('PROPERTY TYPE')\
   .agg(sparkMax(struct(col('CREDIT SCORE'), col('PROPERTY STATE'))).alias('max')).show()
```

```python
df1.groupby('PROPERTY TYPE').agg(sparkMin(struct(col('CREDIT SCORE'), col('PROPERTY STATE'))).alias('min')).show()
```

```python
df_min = df1.select('CREDIT SCORE','PROPERTY STATE')\
          .groupBy(col("PROPERTY STATE")).agg(sparkMin(col("CREDIT SCORE")).alias('min'))
df_min.show()
```

```python
df_max = df1.filter(df1['Credit Score'] != 9999).select('CREDIT SCORE','PROPERTY STATE')\
   .groupBy(col("PROPERTY STATE")).agg(sparkMax(col("CREDIT SCORE")).alias('max'))
df_max.show()
```

```python
df_max.collect()
df_max2 = df_max.collect()
labels = [row['PROPERTY STATE'] for row in df_max2]
counts = [row['max'] for row in df_max2]
fig, ax = plts.subplots(figsize=(15,8))
plts.bar(labels, counts)
```

```python
df_min.collect()
df_min2 = df_min.collect()
labels = [row['PROPERTY STATE'] for row in df_min2]
counts = [row['min'] for row in df_min2]
fig, ax = plts.subplots(figsize=(15,8))
plts.bar(labels, counts)
```
